```python
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
 
y = target 

X = news.drop(['shares'],axis=1)

#normalize the data attributes
normalized_X=preprocessing.normalize(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=52) 
```

```python
#KNN Lab 3
from sklearn.neighbors import KNeighborsClassifier 
neigh = KNeighborsClassifier(n_neighbors=21) 
neigh.fit(X, y) 
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
y_predict = neigh.predict(X_test) 
print(confusion_matrix(y_test, y_predict))
```

```python
print("accuracy= ",accuracy_score(y_test, y_predict)*100)
```

```python
#For KNN above, it did just slightly worse than the other models below.
```

```python
#PCA: Normalized the data
from sklearn.preprocessing import scale
x=scale(X)
```

```python
#Implement PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
PC=pca.fit(x)
```

```python
#PC Components
PC.components_
```

```python
#Variance in components
PC.explained_variance_ratio_
```

```python
#New features for all the instances
New_feats=pca.fit_transform(x)
```

```python
New_feats
```

```python
# Percentage of variance explained for each components
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))
```

```python
#The first two components above explain 83% of the variance which is a big percentage 
```

```python
from sklearn import svm
# Design the model
clf = svm.SVC(kernel='linear')
#clf = svm.LinearSVC(C=1)
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
```
