```python
#Factor Analysis; personality text data
import pandas as pd
dat = pd.read_csv('personality0.txt', sep=" ") 
dat
```

```python
#Correlogram
import matplotlib.pyplot as plt
import numpy as np
corr = dat.corr() 
fig = plt.figure() 
ax = fig.add_subplot(111) 
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1) 
fig.colorbar(cax) 
fig.set_size_inches([16,16]) 
ticks = np.arange(0,len(dat.columns),1) 
ax.set_xticks(ticks) 
plt.xticks(rotation=90) 
ax.set_yticks(ticks) 
ax.set_xticklabels(dat.columns) 
ax.set_yticklabels(dat.columns) 
plt.show()
```

```python
#Factor analysis
from sklearn.preprocessing import scale 
from sklearn.decomposition import FactorAnalysis
data_normal = scale(dat) 
fa = FactorAnalysis(n_components = 10) 
fa = fa.fit(data_normal)
```

```python
feature_names = dat.columns 
factor_names = ['Factor'+str(x) for x in range(1,11)] 
pd.DataFrame(np.transpose(fa.components_), index=feature_names , 
             columns=factor_names)
```
